```python
import pandas as pd
import random
import spotipy
from spotipy.oauth2 import SpotifyClientCredentials
from random import randint
from time import sleep
```

```python
secrets_file = open(r"C:\Users\jasmi\Ironhack\secrets.txt","r")
string = secrets_file.read()
```

```python
secrets_dict={}
for line in string.split('\n'):
    if len(line) > 0:
        secrets_dict[line.split(':')[0]]=line.split(':')[1]
```

```python
sp = spotipy.Spotify(auth_manager=SpotifyClientCredentials(client_id=secrets_dict['cid'],
                                                           client_secret=secrets_dict['csecret']))
```

```python
def get_playlist_tracks(playlist_id):
    results = sp.user_playlist_tracks("spotify",playlist_id)
    tracks = results['items']
    while results['next']!=None:
        results = sp.next(results)
        tracks = tracks + results['items']
        sleep(randint(1,3000)/1000)
    return tracks
```

```python
def flatten(input_list):
    return [item for sublist in input_list for item in sublist]
```

```python
def get_name_artists_from_track(track):
    return [(track["name"],artist["name"],track["uri"]) for artist in track["artists"]]
```

```python
def get_name_artist_from_playlist_item(playlist_item):
    return get_name_artists_from_track(playlist_item['track'])
```

# Playlist 6

```python
playlist6 = get_playlist_tracks("5om77CK9VCES7ymgqtBv9b")
```

```python
playlst_6 = list(map(get_name_artist_from_playlist_item,playlist6))
```

```python
playlst_6 = flatten(playlst_6)
```

```python
df_6 = pd.DataFrame(playlst_6)
```

```python
audio6 = []

for x in df_6[2]:
    audio6.append(sp.audio_features(x))
```

```python
audio6 = flatten(audio6)
audio6_df = pd.DataFrame(audio6)
audio6_df
```

```python
play_6 = pd.concat([df_6 , audio6_df], axis=1)
```
